## 1) Business Understanding

**The Company**

Consider a well-established company operating in the retail food sector. Presently they have around several hundred thousands of registered customers and serve almost one million consumers a year. They sell products from 5 major categories: **wines**, rare **meat** products, exotic **fruits**, specially prepared **fish** and **sweet products**. These can further be divided into  ***gold* and regular products**. The customers can order and acquire products through **3 sales channels**: **physical stores**, **catalogs** and **companyâ€™s website**. Globally, the company had **solid revenues and a healthy bottom line in the past 3 years**, but the **profit growth perspectives for the next 3 years are not promising...** For this reason, several strategic initiatives are being considered to invert this situation. One is to improve the performance of marketing activities, with a special focus on marketing campaigns.


**The Marketing Department**

The marketing department was pressured to **spend its annual budget more wisely**. The CMO perceives the importance of having a more quantitative approach when taking decisions, reason why a small team of data scientists was hired with a clear objective in mind: **to build a predictive model which will support direct marketing initiatives**. Desirably, the success of these activities will prove the value of the approach and convince the more skeptical within the company. 

**The Objective**


 
The objective of the team is to build a predictive model that will produce the highest profit for the next direct marketing campaign, scheduled for the next month. The new campaign, **sixth**, aims at **selling a new gadget to the Customer Database**. To build the model, a **pilot campaign involving 2.240 customers** was carried out. The customers were selected at random and contacted by phone regarding the acquisition of the gadget. During the following months, customers who bought the offer were properly labeled. The **total cost of the sample campaign was 6.720MU** and the **revenue generated by the customers who accepted the offer was 3.674MU**. Globally the campaign had a **profit of -3.046MU**. The **success rate of the campaign was 15%**.

**The objective is of the team is to develop a model that predicts customer behavior and to apply it to the rest of the customer base**. Hopefully the model will allow the company to cherry pick the customers that are most likely to purchase the offer while leaving out the non-respondents, making the next campaign highly profitable. Moreover, other than maximizing the profit of the campaign, the CMO is interested in understanding to study the characteristic features of those customers who are willing to buy the gadget.

**Customers who have accepted offers**

Just checking the number of accepted offers is not enough to understand how many customers have accepted offers, which is going to help us to achieve our main goal here, as a customer could well have accepted offers more than once (there are no rules to avoid this scenario).

Hence, I'm going to create a new column called *HasResponded*, which will be composed the the sum of the features related to the offers acceptance through the campaings, to investigate how many customers have indeed accepted offers, and for those who have accepted more than one (if there are some), how many offers they have accepted:


$$ HasResponded = AcceptedCmp1 + AcceptedCmp2 + AcceptedCmp3 + AcceptedCmp4 +AcceptedCmp5   $$



```
df['HasResponded'] = df['AcceptedCmp1'] + df['AcceptedCmp2'] + df['AcceptedCmp3'] + df['AcceptedCmp4'] + df['AcceptedCmp5']
df['HasResponded'] = df['HasResponded'].astype(int)

#Support Feature
df['AcceptedOffers'] = (np.array(df['HasResponded'] >=1))
```

For all the other categorical features, I'm going to check two main aspects:

*   How's the **customer distribution per category** to have a better feeling of the demographic characteristics of the population we are working with;
*   The **discriminatory power** of each category on that feature. It means, the proportion of customers how have accepted offers on the group of customers that belongs to a certain category.

I'm going to set a **threshold of 0.3** to consider a category relevant on a feature for now. This is going to be helpfull on future steps.



```
# Define deiscriminatory power threshold
disc_pow_threshold = 30
```

```
def print_disc_pow (labels, disc_pow):
  '''
  INPUT:
  labels - List of category labels
  disc_pow - List of discriminatory power per category

  OUTPUT:
  none

  Function prints a list of discriminatory power per category
  '''

  for i in range(len(labels)):
    print('Discriminatory Power of {} : {:.2f} %'.format(labels[i],disc_pow[i]))
```

```
# Education - Discriminatory Power per Category
disc_pow_complain = df.groupby('Complain')['HasResponded'].mean().values*100
disc_pow_complain_labels = df.groupby('Complain')['HasResponded'].mean().index
print_disc_pow (disc_pow_complain_labels, disc_pow_complain)
```

### 3.2 - Quantitative Features

**Customer Enrollment Time**

Now, I'm checking customers enrollment time by looking at the feature *DtCustomer*.

Instead of looking at dates directly, I'm going to convert this feature for the amount of days the customers has been enrolled:

$$ EnrollmentTime: date_{today} - date_{subscription}  $$


```
# Today's date
today = pd.to_datetime(date.today())

# Convert values to datetime format
df['EnrollmentTime'] = pd.to_datetime(df['Dt_Customer'], format='%Y-%m-%d', errors='coerce')
```

```
# Calculate the number of days the user has been using the app

membership_days = []
for i in range(df.shape[0]):
  membership_days.append((today - df['EnrollmentTime'][i]).days)
```

```
#Drop DtCustomer
df.drop(columns='Dt_Customer', inplace=True)

# Update EnrollmentTime
df['EnrollmentTime'] = membership_days

# Distribution plot
plt.figure(figsize = (18,6));
plt.title('Distribution of Enrollment Time');
sns.distplot(df['EnrollmentTime'], color='blue');
```

### 4.2 K-Means Clustering

One important hyperparameter for the K-Means Clustering method is the number of clusters *k*. One way to tune this hyperparameter and get an optimal model is the **Elbow Rule**, that consists in trying different values of *k* and measure the withim cluster sum-of-squares, also called model's inertia for each given *k*. Then, by ploting the inertia x *k*, it's possible to visually determine the optimal value of clusters, which is the point where the inertia values start to decrease in a linear fashion.

```
def elbow_rule(K, data):
  '''
  INPUT
  k - Number of clusters
  data - Input data

  OUTPUT
  inertia_values - List of model's inertia for each value in the range (1,k+1)
  '''

  cluster_values = list(range(1,K+1))
  inertia_values = []

  for num_k in range(1,K+1):
    model = KMeans(n_clusters= num_k, init='k-means++', max_iter=500, random_state = 11)
    model.fit(data)
    inertia_values.append(model.inertia_)

  return inertia_values
```

```
# Hyperparemeter optimization - Elbow Rule for k = 20
kmeans_results = elbow_rule(20,X_seg_scaled)
```

```
# ELbow Plot for visual analysis
fig = go.Figure()

fig.add_trace(go.Scatter(x=list(range(1,21)),
                         y=kmeans_results,
                         mode='lines+markers'))

fig.update_layout(xaxis = dict(
    tickmode = 'linear',
    tick0 = 1,
    dtick = 1),
    title_text="Elbow Rule - Within Cluster Sum of Squared Distances x K Values",
                  xaxis_title="K values",
                  yaxis_title="Cluster sum of squared distances")

fig.show()
```

By analyzing the figure above, it's possible to notice that model's inertia starts to decrease in a smoother way after **k=8**, almost in a linear fashion. Hence, I'm defining **8 as the optimal number of clusters** for the customer segmentation.

```
# Optimal clustering model with k = 6
optimal_kmeans_model = KMeans(n_clusters= 8, init='k-means++', max_iter=500, random_state = 11)
optimal_kmeans_model.fit(X_seg_scaled)
clusterized_data = optimal_kmeans_model.predict(X_seg_scaled)
```

```
#Education
rel_Education = X['Education'].apply(lambda x: 'Other' if (x != 'Graduation') & (x != 'PhD') else x )

#Marital Status
rel_Marital = X['Marital_Status'].apply(lambda x: 'Other' if ((x != 'Absurd') & (x != 'Alone') & (x != 'Widow') & (x != 'Married'))  else x )

#Kidhome
rel_Kidhome = X['Kidhome'].apply(lambda x: 0 if (x!= 0) else 1)

#Teenhome
rel_Teenhome = X['Teenhome'].apply(lambda x: 1 if (x!= 1) else 0)

```

Now that only the less relevant categories of each feature were removed, I'm going to create dummy variables for *Education* and *Marital_Status* that have more then 2 categories each:

```
# Education - Dummy Variables
Education_dummies = pd.get_dummies(rel_Education, prefix='Education', drop_first=False)
# Marital Status - Dummy Variables
Marital_dummies = pd.get_dummies(rel_Marital, prefix='Marital', drop_first=False)
# Kidhome
Kidhome_dummies = pd.get_dummies(rel_Kidhome, prefix='Kidhome', drop_first=True)
# Teenhome
Teenhome_dummies = pd.get_dummies(rel_Teenhome, prefix='Teenhome', drop_first=True)
```

```
 # Concatenate all the dummie variables together
X_cat = pd.concat([Education_dummies,Marital_dummies,Kidhome_dummies,Teenhome_dummies], axis=1)
```

#### 5.2.2 - Numerical Features

Now, it's time to deal with X's numerical features. As it was the case on the unsupervised machine learning algorithm, all features here have different ranges, which is not good for our machine learning algorithm.

Then, I'm going to follow the same procedure I did during the clustering step, and apply a MinMax Scaler on the data, then all of the features will be contained in the range [0,1].

```
# Drop categorical features from X
X.drop(['Education','Marital_Status','Kidhome','Teenhome'], axis=1, inplace=True)

# build scaler
scaler = MinMaxScaler()
# fit data
scaler.fit(X)
#transform data
X_num_scaled = scaler.transform(X)

# Create a data frame with the scaled data
X_num = pd.DataFrame(data=X_num_scaled, columns=X.columns)
X_num.reset_index(inplace=True)
```

Then, now it's possible to merge all the proprocessed data into *X*:

```
# Concatenate dataframes
X = pd.concat([X_num, X_cat], axis=1)
X.drop(['index'], axis=1, inplace=True)

# Check X
X.head()
```

### 5.3 - Modeling

First step here, is to split the data into two sets: **Training** and **Test** sets.

I'm going to use 10% of the data as **Test** dataset:

```
# Split data into traning and test sets
np.random.seed(11)
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.1, random_state=11)
```

One's facing a classification problem here, where the model needs to predict if the customer is going to accept the offer or not [1 or 0].

I'm going to try two different classification models, and then compare their performances according to the model's **accuracy**.

They are:

*   **Logistic Regression**
*   **Decision Tree**

As both model's performance are dependent of the choices of Hyperparameters, I'm going to use GridSearchCV to automate the hyperparameter testing, and optimize our classification models.

All the information regarding the parameters of each model is available at the Sklearn documentation:

[Logistic Regression documentation](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html)

[Decision Tree Classifier documentation](https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeClassifier.html)




#### 5.3.1 - Logistic Regression

**Hyperparameter Optimization**

```
# Define the Machine Learning Pipeline

lr_pipeline = Pipeline([
                        ('clf', LogisticRegression())
                        ])
```

```
# Check LogisticRegression Hyperparameters
lr_pipeline.get_params()
```

```
# Define hyperparameters to be optimized in the GridSearchCV
lr_parameters = {
    'clf__penalty': ('l1','l2','elasticnet','none'),
    'clf__solver': ('lbfgs','newton-cg','liblinear','sag','saga'),
    'clf__max_iter': (100,200,500)      
                }
```

```
# Optimize and Fit
cv_lr = GridSearchCV(lr_pipeline, param_grid=lr_parameters, verbose=2, cv=5)
cv_lr.fit(X_train, y_train)
```

```
# Optimized Hyperparameters Logistic Regression Model
optimized_model_lr = cv_lr.best_estimator_
print(optimized_model_lr)
```

**Model Evalutation**

```
# Predictions using the test dataset
y_hat_lr = optimized_model_lr.predict(X_test)

# Model's Prediction Accuracy
acc_lr = accuracy_score(y_test, y_hat_lr)*100
print('Logistic Regression Model Accuracy : {:.4f} % '.format(acc_lr))
```

```
# Model evaluation - Precision to True Positives
print(classification_report(y_test, y_hat_lr))
```

#### 5.3.2 - Decision Tree

```
# Define the Machine Learning Pipeline

dt_pipeline = Pipeline([
                        ('clf', DecisionTreeClassifier())
                        ])
```

```
# Check DecisionTreeClassifier Hyperparameters
dt_pipeline.get_params()
```

```
# Define hyperparameters to be optimized in the GridSearchCV
dt_parameters = {
    'clf__criterion': ('gini','entropy'),
    'clf__splitter': ('best','random'),
    'clf__min_samples_split': (2,4,6,8),
    'clf__min_samples_leaf': (1,2,4),   
                }
```

```
# Optimize and Fit
cv_dt = GridSearchCV(dt_pipeline, param_grid=dt_parameters, verbose=2, cv=5)
cv_dt.fit(X_train, y_train)
```

```
# Optimized Hyperparameters Decision Tree Model
optimized_model_dt = cv_dt.best_estimator_
print(optimized_model_dt)
```

```
# Predictions using the test dataset
y_hat_dt = optimized_model_dt.predict(X_test)

# Model's Prediction Accuracy
acc_dt = accuracy_score(y_test, y_hat_dt)*100
print('Decision Tree Model Accuracy : {:.4f} % '.format(acc_dt))
```

```
# Model evaluation -Precision to True Positives
print(classification_report(y_test, y_hat_dt))
```
